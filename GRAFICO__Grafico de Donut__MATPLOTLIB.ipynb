```python
# Gráfico de Donut utilizando a biblioteca Matplotlib.

import matplotlib.pyplot as plt

fig, ax = plt.subplots()
circle = plt.Circle((0, 0),0.64, color="white")
lbls = ["Custo 1", "Custo 2", "Custo 3", "Custo 4", "Custo 5"]
ax.pie([20,20,20,20,20],
       labels=lbls,
       autopct="%1.1f%%",
       pctdistance=.82) 
ax.add_artist(circle)
ax.set_title("Gráfico de Donut", fontsize=16)
plt.show()
```
