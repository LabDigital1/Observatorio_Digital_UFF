```python
# Importar bibliotecas
import datetime as dt 
import seaborn as sns
from pandas_datareader import data as pdr 
import matplotlib.pyplot as plt 

data_final = dt.datetime.now()
data_inicial = data_final - dt.timedelta(days=300)

# Buscar as ações necessárias utilizando o Yahoo! Finance 
precos = pdr.get_data_yahoo(['PETR4.SA', 'VALE3.SA', 'COCA34.SA', 'MGLU3.SA'], data_inicial, data_final)['Adj Close']

# Calcular o retorno diário das ações
retornos = precos.pct_change().dropna()

# Criar a Matriz de Correlação
correlacao = retornos.corr()
plot = sns.heatmap(correlacao, annot=True, fmt='.2f', linewidths=.6)
plt.show()
```
