```python
import fundamentus
import pandas as pd
import matplotlib.pyplot as plt

# Construir um DataFrame
df = fundamentus.get_resultado()

# Realizar a filtragem do dado
filtro =  df[ (df.pl < 10) & (df.mrgebit > 0.2)& (df.dy > 0.1375)]

filtro.sort_values('dy', ascending = False )

# Montar o gráfico
plt.figure(figsize = (20, 6));
plt.bar(filtro.index, filtro.dy);
plt.xticks(rotation = 45);
plt.title('Ações da B3 Pagando Dividendos Acima da Taxa SELIC, com PL Abaixo de 10 e Margem EBTIDA Acima de 20%')
```
