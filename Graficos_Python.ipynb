```python
# Gráfico de barras para Múltiplas Variáveis

import numpy as np 
import matplotlib.pyplot as plt 
  
X = ['1o Ano','2o Ano','3o Ano','4o Ano']
Ymanha = [15,25,16,19]
Ztarde = [17,14,21,26]
  
X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2, Ymanha, 0.4, label = 'Manhã')
plt.bar(X_axis + 0.2, Ztarde, 0.4, label = 'Tarde')
  
plt.xticks(X_axis, X)
plt.xlabel("Turnos")
plt.ylabel("Number de Alunos pr Turno")
plt.title("Número de Alunos em Cada Série")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
circle = plt.Circle((0, 0),0.64, color="white")
lbls = ["Energia", "Salários", "Fornecedores", "Marketing", "Combustível"]

ax.pie([8,34,28,7,16],
       labels=lbls,
       autopct="%1.1f%%",
       pctdistance=.82)
ax.add_artist(circle)
ax.set_title("Gráfico de Donut", fontsize=16)
plt.show()
```
