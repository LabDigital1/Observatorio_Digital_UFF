```python
# Importar bibliotecas
import matplotlib.pyplot as plt
import numpy as np

# Definir os dados
y = np.array([30, 15, 10, 25])

# Definir as classificações
cargos = ['Administrativo', 'Financeiros', 'Marketing', 'Pessoas']

# Dar ênfase ao espaço determinado 
espaco = [0.2, 0, 0, 0]

# Construir o gráfico
plt.title('Gastos por área')
plt.pie(y, labels = cargos, explode = espaco, shadow = True)
plt.show()
```
