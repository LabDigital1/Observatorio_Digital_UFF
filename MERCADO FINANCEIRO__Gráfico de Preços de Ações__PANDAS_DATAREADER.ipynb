```python
# Gráfico dos preços de fechamento de um portfólio de ações ao longo do tempo utilizando as bibliotecas matplotlib e yfinance.

import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt

tickers = ["VALE3.SA", "USIM5.SA", "GGBR4.SA", "CSNA3.SA"]
stock_prices_br = pd.DataFrame()
for t in tickers:
	 stock_prices_br[t] = wb.DataReader(t, data_source = "yahoo", start ="2020-1-1")["Adj Close"]

stock_prices_br.plot(figsize = (15, 6))
plt.show
```
