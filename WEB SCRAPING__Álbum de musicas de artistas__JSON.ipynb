```python
# Programa II: 

# Baixar os dados da API iTunes sobre as músicas de um determinado grupo ou artista e gerar uma planilha Excel com esses dados:

# Importar as Bibliotecas
import pandas as pd
import requests
import json

# Buscar o site onde iremos retirar as informações
base_site = "https://itunes.apple.com/search"

# Definir os dados que iremos buscar
r = requests.get(base_site, params = {"term": "Anitta", "country": "us"})
info = r.json()
songs_df = pd.DataFrame(info["results"])

# Nomear a planilha
songs_df.to_excel("musicas_anitta_info.xlsx")
songs_df
```
