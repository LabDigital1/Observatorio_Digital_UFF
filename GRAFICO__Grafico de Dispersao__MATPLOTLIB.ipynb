```python
# Gráfico de dispersão criado com a biblioteca Matplotlib

# Importando a biblioteca
import matplotlib.pyplot as plt

# Inserindo os dados
x = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
y = [5, 6, 4, 6, 4, 7, 10, 11, 11, 5, 7, 6]
 
# Definindo o tipo e a cor dos dados   
plt.scatter(x, y, label = 'Pontos', color = 'b', marker = '.', s = 100)

# Dando nome as variáveis e um titulo para o gráfico
plt.legend('Quantidade vendida/Mês')
plt.xlabel("Meses")
plt.ylabel("Qtde Vendida (em milhares)")

# Medindo a distancia entre os pontos e o eixo
plt.xlim(left=-1)
plt.ylim(bottom=2)
  
plt.show()
```
