```python
# Importando bibliotecas
import pandas as pd
from pytrends.request import TrendReq

# Definindo o fuso horrio da regio da busca
pytrends = TrendReq(hl='pt-BR', tz=360)

# Especificando termo da busca
pytrends.build_payload(kw_list=['copa do mundo'])

df = pytrends.interest_by_region() 
df.sort_values(by=['copa do mundo'], ascending=False) 


# Plotando frequncia de buscas no google em cada pas
df.reset_index().plot(x='geoName', y='copa do mundo', figsize=(120, 10), kind='bar')
```
